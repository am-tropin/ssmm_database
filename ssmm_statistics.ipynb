```python
import pandas as pd
```

```python
ssmm_candidates_x_votes_info_conv_1_11 = pd.read_csv('ssmm_candidates_x_votes_info_conv_1_11.csv', sep='\t', encoding='utf-8')
ssmm_candidates_x_votes_info_conv_1_11.head()
```

```python
ssmm_organizers_info_upd = pd.read_csv('ssmm_organizers_info_upd.csv', sep='\t', encoding='utf-8')
ssmm_organizers_info_upd.head()
```

```python
academic_group_stat = ssmm_candidates_x_votes_info_conv_1_11.groupby(['year_stream_no','academic_group_no']).agg({'conv_x_name':'count'}).reset_index().rename(columns={'conv_x_name':'application_cnt'}).sort_values(['application_cnt'],
              ascending = [False]).reset_index().drop(['index'], axis=1)
academic_group_stat[-30:]

# 107 раньше была в 1.2, теперь в 1.1
# 113 и 114 раньше не было
```

```python
year_stream_stat = ssmm_candidates_x_votes_info_conv_1_11.groupby(['year_stream_no']).agg({'conv_x_name':'count'}).reset_index().rename(columns={'conv_x_name':'application_cnt'}).sort_values(['application_cnt'],
              ascending = [False]).reset_index().drop(['index'], axis=1)
year_stream_stat[:60]
```

```python
ssmm_organizers_info_final = ssmm_organizers_info_upd[ssmm_organizers_info_upd['organizer_name'] != None].groupby(['organizer_name']).agg({'year_x_year_stream':'count'}).reset_index().rename(columns={'year_x_year_stream':'elections_cnt'}).sort_values(['elections_cnt'],
              ascending = [False]).reset_index().drop(['index'], axis=1)
ssmm_organizers_info_final.head(20)
```
