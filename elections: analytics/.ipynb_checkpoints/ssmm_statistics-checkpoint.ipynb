```python
ssmm_elections_candidates_1_11 = pd.read_csv('../library: main/ssmm_elections_candidates_1_11.csv', sep='\t', encoding='utf-8')
ssmm_elections_candidates_1_11.head(8)
```

```python
ssmm_elections_streams_1_11 = pd.read_csv('../library: main/ssmm_elections_streams_1_11.csv', sep='\t', encoding='utf-8')
ssmm_elections_streams_1_11.head()
```

```python
ssmm_programs_info_conv_1_11 = pd.read_csv('../library: main/ssmm_programs_info_conv_1_11.csv', sep='\t', encoding='utf-8')
ssmm_programs_info_conv_1_11.head()
```

```python
'''
ФЛАГИ ДЛЯ АНАЛИТИКИ ssmm_elections_streams_1_11:
- добавить флаг превышения квоты по ТА
- добавить флаг наличия конкуренции на потоке
'''
```

```python
# конкуренция на потоке

temp_df = ssmm_elections_streams_1_11[['convocation_no','year_stream_no','quota','applications_cnt','elected_cnt']]

competition_df = temp_df[temp_df['applications_cnt'] > temp_df['quota']].reset_index().drop(['index'],axis=1)
competition_df = competition_df.drop_duplicates().reset_index().drop(['index'],axis=1)
competition_df
```

```python
competition_df.to_csv('ssmm_streams_with_competition.csv', sep='\t', encoding='utf-8', index=False)
# ssmm_organizers_info_upd.to_csv('../library: main/ssmm_organizers_1_11.csv', sep='\t', encoding='utf-8', index=False)
```

```python
# нужны потоки с хотя бы одним неизбранным:

with_loser_df = temp_df[temp_df['applications_cnt'] > temp_df['elected_cnt']].reset_index().drop(['index'],axis=1)
with_loser_df = with_loser_df.drop_duplicates().reset_index().drop(['index'],axis=1)
with_loser_df
```

```python
branches_list = list(ssmm_programs_info_conv_1_11)[-11:]
branches_list
```

```python
# ВНУТРИ ПОТОКА ПОДТЯНУЛИСЬ НЕ ВСЕ КАНДИДАТЫ, потому что не всех были программы

competition_programs_1 = ssmm_programs_info_conv_1_11.merge(ssmm_elections_candidates_1_11, left_on='conv_x_name', right_on='conv_x_name', how='inner')
# competition_programs_1.head(8)
competition_programs_2 = competition_programs_1.merge(with_loser_df, how='inner')
competition_programs_3 = competition_programs_2.merge(ssmm_elections_streams_1_11, left_on=['convocation_no', 'year_stream_no'], right_on=['convocation_no', 'year_stream_no'], how='left')
competition_programs_3 = competition_programs_3[['conv_x_name','convocation_no','year_no','elected_flg']+branches_list]
# 'year_stream_no'
competition_programs_3 #.head(30)

# добавить пол кандидата?
```
