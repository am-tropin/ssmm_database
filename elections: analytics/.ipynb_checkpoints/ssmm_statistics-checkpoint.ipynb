```python
import pandas as pd
```

```python
ssmm_candidates_x_votes_info_conv_1_11 = pd.read_csv('../library: main/ssmm_elections_candidates_1_11.csv', sep='\t', encoding='utf-8')
ssmm_candidates_x_votes_info_conv_1_11.head()
```

```python
ssmm_organizers_info_upd = pd.read_csv('ssmm_organizers_info_upd.csv', sep='\t', encoding='utf-8')
ssmm_organizers_info_upd.head()
```

```python
academic_group_stat = ssmm_candidates_x_votes_info_conv_1_11.groupby(['year_stream_no','academic_group_no']).agg({'conv_x_name':'count'}).reset_index().rename(columns={'conv_x_name':'application_cnt'}).sort_values(['application_cnt'],
              ascending = [False]).reset_index().drop(['index'], axis=1)
academic_group_stat[-30:]

# 107 раньше была в 1.2, теперь в 1.1
# 113 и 114 раньше не было
```

```python
year_stream_stat = ssmm_candidates_x_votes_info_conv_1_11.groupby(['year_stream_no']).agg({'conv_x_name':'count'}).reset_index().rename(columns={'conv_x_name':'application_cnt'}).sort_values(['application_cnt'],
              ascending = [False]).reset_index().drop(['index'], axis=1)
year_stream_stat[:60]
```

```python
ssmm_organizers_info_final = ssmm_organizers_info_upd[ssmm_organizers_info_upd['organizer_name'] != None].groupby(['organizer_name']).agg({'year_x_year_stream':'count'}).reset_index().rename(columns={'year_x_year_stream':'elections_cnt'}).sort_values(['elections_cnt'],
              ascending = [False]).reset_index().drop(['index'], axis=1)
ssmm_organizers_info_final.head(20)
```

```python
ssmm_elections_candidates_1_11 = pd.read_csv('../library: main/ssmm_elections_candidates_1_11.csv', sep='\t', encoding='utf-8')
ssmm_elections_candidates_1_11.head(8)
```

```python
ssmm_elections_streams_1_11 = pd.read_csv('../library: main/ssmm_elections_streams_1_11.csv', sep='\t', encoding='utf-8')
ssmm_elections_streams_1_11.head()
```

```python
ssmm_programs_info_conv_1_11 = pd.read_csv('../library: main/ssmm_programs_info_conv_1_11.csv', sep='\t', encoding='utf-8')
ssmm_programs_info_conv_1_11.head()
```

```python
'''
ФЛАГИ ДЛЯ АНАЛИТИКИ ssmm_elections_streams_1_11:
- добавить флаг превышения квоты по ТА
- добавить флаг наличия конкуренции на потоке
'''
```

```python
# конкуренция на потоке

temp_df = ssmm_elections_streams_1_11[['convocation_no','year_stream_no','quota','applications_cnt','elected_cnt']]

competition_df = temp_df[temp_df['applications_cnt'] > temp_df['quota']].reset_index().drop(['index'],axis=1)
competition_df = competition_df.drop_duplicates().reset_index().drop(['index'],axis=1)
competition_df
```

```python
competition_df.to_csv('ssmm_streams_with_competition.csv', sep='\t', encoding='utf-8', index=False)
# ssmm_organizers_info_upd.to_csv('../library: main/ssmm_organizers_1_11.csv', sep='\t', encoding='utf-8', index=False)
```

```python
# нужны потоки с хотя бы одним неизбранным:

with_loser_df = temp_df[temp_df['applications_cnt'] > temp_df['elected_cnt']].reset_index().drop(['index'],axis=1)
with_loser_df = with_loser_df.drop_duplicates().reset_index().drop(['index'],axis=1)
with_loser_df
```

```python
branches_list = list(ssmm_programs_info_conv_1_11)[-11:]
branches_list
```

```python
# НЕКОРРЕКТНЫЙ ДЖОЙН, НЕ ВСЕ КАНДИДАТЫ ВНУТРИ ПОТОКА ПОДТЯНУЛИСЬ

competition_programs_1 = ssmm_programs_info_conv_1_11.merge(ssmm_elections_candidates_1_11, how='inner')
competition_programs_2 = competition_programs_1.merge(with_loser_df, how='inner')
competition_programs_2 = competition_programs_2[['conv_x_name','year_stream_no','elected_flg']+branches_list]
competition_programs_2.head(8)
```

```python
# Тен-Акушевич

TA_df = temp_df[temp_df['elected_cnt'] > temp_df['quota']].reset_index().drop(['index'],axis=1)
TA_df = TA_df.drop_duplicates().reset_index().drop(['index'],axis=1)
TA_df
```
