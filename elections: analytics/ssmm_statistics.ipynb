```python
counter_df = ssmm_elections_candidates_1_11.groupby(['convocation_no','year_stream_no']).agg({'conv_x_name':'count'}).rename(columns={'conv_x_name':'applications_cnt'})
counter_df = counter_df.reset_index()
counter_df
```

```python
# сделать формулу расчета квоты от потока (до 5 курса - фикс, с 6 курса - от численности потока)

def def_quota(convocation_no, year_stream_no, stream_size):
    if convocation_no <= 5:
        if 'm' in year_stream_no:
            return 1
        elif int(year_stream_no[-1]) == 4:
            return 1
        else:
            return 3
    else:
        return min(stream_size//25 + 1, 3)
    
def_quota(6,'m1',30)
```

```python
quota_df = ssmm_elections_streams_1_11[['convocation_no', 'year_stream_no', 'stream_size', 'voted_cnt']]
quota_df['quota'] = quota_df.apply(lambda x: def_quota(x['convocation_no'], x['year_stream_no'], x['stream_size']), axis=1).astype(int)
quota_df
```

```python
# quota_df[quota_df['quota'].isna()]
```

```python
full_df = quota_df.merge(counter_df,how='outer')
full_df['applications_cnt'] = full_df['applications_cnt'].fillna(0).astype(int)
full_df
```

```python
competition_df = full_df[full_df['applications_cnt'] > full_df['quota']].reset_index().drop(['index'],axis=1)
competition_df
```

```python
competition_df.to_csv('ssmm_streams_with_competition.csv', sep='\t', encoding='utf-8', index=False)
# ssmm_organizers_info_upd.to_csv('../library: main/ssmm_organizers_1_11.csv', sep='\t', encoding='utf-8', index=False)
```
