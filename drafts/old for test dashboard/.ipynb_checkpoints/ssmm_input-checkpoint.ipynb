```python
mm_input = pd.read_excel('mm_input.xlsx', sheet_name = 'mm_input_nominations')
mm_input.head()
```

```python
# справочник по номерам курсов и созывов 
# - ЛУЧШЕ БЫ АВТОМАТИЗИРОВАТЬ

mm_year_stream = pd.read_excel('mm_dictionaries.xlsx', sheet_name = 'year_stream_info')
mm_year_stream = mm_year_stream[['year_stream_no','year_no','stream_no']]
mm_year_stream = mm_year_stream[(mm_year_stream.year_no != 'м1') & (mm_year_stream.year_no != 'м2')]
mm_year_stream = mm_year_stream.astype({"year_no": int}) # str
mm_year_stream = mm_year_stream.astype({"stream_no": int})
mm_year_stream.head()
```

```python
# ТОЛЬКО НУЖНЫЕ КОЛОНКИ ПРО ВЫДВИЖЕНИЕ

t_nominated = mm_input[['ssmm_id', 'convocation_no', 'iteration_no', 'name', 'year_stream_no', 'votes_cnt', 'mandate_flg']]
t_nominated.head(5)
```

```python
t_nominated['no'] = np.arange(1,len(t_nominated)+1)
t_nominated.head(5)
```

```python
t_nominated['year_stream_no'].astype(str)
# t_nominated['lower_name'] = t_nominated['name'].str.lower()
t_nominated['lower_name'] = t_nominated['name'].str.title()
t_nominated.head()
```

```python
# + нормальные номера курса и потока

t_nominated_y = pd.merge(t_nominated, mm_year_stream, how='left')
t_nominated_y.head()
```

```python
t_nominated_y.nunique()
```

```python
# ТОЛЬКО ИЗБРАННЫЕ

t_elected = t_nominated_y[t_nominated_y.mandate_flg == 1]
# t_elected = t_elected.astype({"year_no": int})
# t_elected['year_no_str'] = t_elected.year_no
# t_elected = t_elected.astype({"year_no": str})
t_elected.head()
```

```python
# СДЕЛАТЬ ФЛАГ для студентов, которые поступили раньше года создания ССММ

# t_nominated_y_conv = t_nominated_y[t_nominated_y.convocation_no >= t_nominated_y.year_no]
# t_nominated_y_conv.head()

```

```python
# какой номер курса был у выпускного (в зависимости от номера созыва)
def max_year_no(convocation_no):
    if convocation_no <= 5:
        return 5
    else:
        return 6
    
# флаг выпускного курса в данном созыве
def graduate_year_flg(year_no, convocation_no):
    if year_no == max_year_no(convocation_no):
#         print(year_no)
#         print(max_year_no(convocation_no))
        return 1
    else:
        return 0
```

```python

# ИССЛЕДОВАНИЕ 1: НА КАКОМ КУРСЕ ВПЕРВЫЕ ВЫДВИГАЮТСЯ В ССММ
```

```python
# поправка на первые созывы: удаляем студентов, которые поступили раньше года создания ССММ

t_nominated_y_conv = t_nominated_y[t_nominated_y.convocation_no >= t_nominated_y.year_no]
t_nominated_y_conv.head()
```

```python
t_first_nomination = t_nominated_y_conv[['lower_name','year_no']].groupby(['lower_name'], as_index = False).agg({'year_no':'min'}).rename(columns={'year_no':'year_of_first_nomination'}).sort_values(by=['year_of_first_nomination'], ascending = False)
t_first_nomination.head(13)
```

```python
t_first_year_nomination = t_first_nomination[['year_of_first_nomination','lower_name']].groupby(['year_of_first_nomination'], as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['year_of_first_nomination'], ascending = False)
t_first_year_nomination.head(10)
```

```python

# ИССЛЕДОВАНИЕ 2: КАНДИДАТЫ НА ВЫПУСКНОМ КУРСЕ
```

```python
t_graduated = t_nominated_y[['lower_name','convocation_no','year_no']][((t_nominated_y.year_no == 5) & (t_nominated_y.convocation_no <= 5)) | ((t_nominated_y.year_no == 6) & (t_nominated_y.convocation_no > 5))]
t_graduated['graduate_year_flg'] = 1
t_graduated.head()
```

```python

# ИССЛЕДОВАНИЕ 3: ОТ КАКОГО ПОТОКА БОЛЬШЕ КАНДИДАТОВ?
```

```python
t_streams = t_nominated_y[['stream_no','lower_name']].groupby(['stream_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['stream_no'], ascending = True)
t_streams.head()
```

```python
# то же, но курс+поток

t_year_streams = t_nominated_y[['year_no','stream_no','lower_name']].groupby(['year_no','stream_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['year_no','stream_no'], ascending = True)
t_year_streams.head()
```

```python
t_year_streams = t_year_streams.sort_values(by=['lower_name_cnt'], ascending = False)
t_year_streams.head()
```

```python

# МЕТРИКА 1: ПОПУЛЯРНОСТЬ СОЗЫВА (число выдвижений в каждый созыв)
```

```python
t_popular_convocations = t_nominated_y[['convocation_no','lower_name']].groupby(['convocation_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['convocation_no'], ascending = True)
t_popular_convocations = t_popular_convocations.rename(columns={'lower_name_cnt':'nomination_cnt'})
t_popular_convocations
```

```python
# t_popular_convocations = t_popular_convocations.sort_values(by=['lower_name_cnt'], ascending = False)
# t_popular_convocations
```

```python

# МЕТРИКА 2: РАЗМЕР СОЗЫВА
# - нет ли бага во 2 созыве? (руками насчитал 36)
```

```python
t_convocation_size = t_elected[['convocation_no','lower_name']].groupby(['convocation_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['convocation_no'], ascending = True)
t_convocation_size = t_convocation_size.rename(columns={'lower_name_cnt':'member_cnt'})
t_convocation_size
```

```python

# МЕТРИКА 3.1: НОВИЧКИ В ЗАЯВКАХ В СОЗЫВ
# - исключая первый созыв (там все были новичками)
```

```python
t_first_nomination = t_nominated_y[['convocation_no','lower_name']].groupby(['lower_name'],as_index = False).agg({'convocation_no':'min'}).rename(columns={'convocation_no':'first_convocation_no'}).sort_values(by=['first_convocation_no'], ascending = True)
t_first_nomination['first_nomination_flg'] = 1
t_first_nomination.head()
```

```python
t_nomination_newbie = t_first_nomination[['first_convocation_no','lower_name']].groupby(['first_convocation_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['first_convocation_no'], ascending = True)
t_nomination_newbie = t_nomination_newbie.rename(columns={'lower_name_cnt':'newbie_nomination_cnt'})
t_nomination_newbie
```

```python
# то же в процентах от числа заявок

t_nomination_newbie_ratio = pd.merge(t_nomination_newbie, t_popular_convocations, how='left', left_on=['first_convocation_no'], right_on = ['convocation_no']) 
# t_nomination_newbie_ratio
# t_nomination_newbie_ratio1 = t_nomination_newbie_ratio[['first_convocation_no','lower_name_cnt_x','lower_name_cnt_y']].rename(columns={'lower_name_cnt_x':'newbie_cnt', 'lower_name_cnt_y':'convocation_start_size'})
t_nomination_newbie_ratio['newbie_nomination_ratio'] = t_nomination_newbie_ratio.newbie_nomination_cnt / t_nomination_newbie_ratio.nomination_cnt
t_nomination_newbie_ratio
```

```python

# МЕТРИКА 3.2: НОВИЧКИ В СОЗЫВЕ
# - исключая первый созыв (там все были новичками)
```

```python
t_first_convocation = t_elected[['convocation_no','lower_name']].groupby(['lower_name'],as_index = False).agg({'convocation_no':'min'}).rename(columns={'convocation_no':'first_convocation_no'}).sort_values(by=['first_convocation_no'], ascending = True)
t_first_convocation['first_convocation_flg'] = 1
t_first_convocation.head()
```

```python
t_convocation_newbie = t_first_convocation[['first_convocation_no','lower_name']].groupby(['first_convocation_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['first_convocation_no'], ascending = True)
t_convocation_newbie
```

```python
# то же в процентах от размера созыва

t_convocation_newbie_ratio = pd.merge(t_convocation_newbie, t_convocation_size, how='left', left_on=['first_convocation_no'], right_on = ['convocation_no']) 
# t_convocation_newbie_ratio
t_convocation_newbie_ratio1 = t_convocation_newbie_ratio[['first_convocation_no','lower_name_cnt','member_cnt']].rename(columns={'lower_name_cnt':'newbie_cnt'})
t_convocation_newbie_ratio1['newbie_ratio'] = t_convocation_newbie_ratio1.newbie_cnt / t_convocation_newbie_ratio1.member_cnt
t_convocation_newbie_ratio1
```

```python

# МЕТРИКА 4: НЕВОЗВРАЩЕНЦЫ ПОСЛЕ ДАННОГО СОЗЫВА
# - вычли выпускников
# - не учитывать последний на текущий момент созыв
# - на будущее: сделать поправку на возвраты спустя созыв (но это редкий кейс, пока забить)
```

```python
# тупиковый подсчет, без учета выпускников

t_last_convocation = t_elected[['convocation_no','lower_name']].groupby(['lower_name'],as_index = False).agg({'convocation_no':'max'}).rename(columns={'convocation_no':'last_convocation_no'}).sort_values(by=['last_convocation_no'], ascending = True)
t_last_convocation['last_convocation_flg'] = 1
t_last_convocation.head()
```

```python
t_elected_2 = pd.merge(t_elected, t_graduated, how='left')
t_elected_not_graduated = t_elected_2[t_elected_2.graduate_year_flg != 1]
t_elected_not_graduated.head()
```

```python
t_last_convocation_not_graduated = t_elected_not_graduated[['convocation_no','lower_name']].groupby(['lower_name'],as_index = False).agg({'convocation_no':'max'}).rename(columns={'convocation_no':'last_convocation_no'}).sort_values(by=['last_convocation_no'], ascending = False)
t_last_convocation_not_graduated.head()
```

```python
t_convocation_exodus = t_last_convocation_not_graduated[['last_convocation_no','lower_name']].groupby(['last_convocation_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['last_convocation_no'], ascending = True)
t_convocation_exodus
```

```python
# то же в процентах от размера созыва
# - БАГ В ПОСЛЕДНЕМ СОЗЫВЕ (ДОЛЖНО БЫТЬ 100%, но не учтены МАГИСТРЫ)

t_convocation_exodus_ratio = pd.merge(t_convocation_exodus, t_convocation_size, how='left', left_on=['last_convocation_no'], right_on = ['convocation_no']) 
# t_convocation_exodus_ratio
t_convocation_exodus_ratio1 = t_convocation_exodus_ratio[['last_convocation_no','lower_name_cnt','member_cnt']].rename(columns={'lower_name_cnt':'exodus_cnt'})
t_convocation_exodus_ratio1['exodus_ratio'] = t_convocation_exodus_ratio1.exodus_cnt / t_convocation_exodus_ratio1.member_cnt
t_convocation_exodus_ratio1
```

```python
# t_convocation_exodus_ratio1.sort_values(by=['exodus_ratio'], ascending = False)
```

```python

# МЕТРИКА 5.2: СРЕДНИЙ НОМЕР КУРСА (среди избранных)
# - научиться конвертировать колонку из str в int
# - посчитать ОТКЛОНЕНИЕ от среднего номера курса по факультету (изи: 3 или 3.5, хард: просуммировать численность по потокам)
```

```python
# t_elected['year_no_int'] = int(t_elected.year_no)
t_avg_year_no = t_elected[['convocation_no','year_no']].groupby(['convocation_no'],as_index = False).agg({'year_no':'mean'}).rename(columns={'year_no':'avg_year_no'}).sort_values(by=['convocation_no'], ascending = True)
t_avg_year_no['avg_year_no_faculty'] = t_avg_year_no.convocation_no.apply(lambda x: (3 if x <= 5 else 3.5))
t_avg_year_no['avg_year_no_diff'] = t_avg_year_no.avg_year_no - t_avg_year_no.avg_year_no_faculty

t_avg_year_no
```

```python

# МЕТРИКА 6: НОВИЧКИ-НЕВОЗВРАЩЕНЦЫ (сколько их выпало на каждый созыв)
# - точное число
# - в долях от всех новичков данного созыва
```

```python
t_convocation_newbie_exodus = pd.merge(t_first_convocation, t_last_convocation_not_graduated, how='inner', left_on=['first_convocation_no','lower_name'], right_on = ['last_convocation_no','lower_name']) 
# t_convocation_newbie_exodus.head()
t_convocation_newbie_exodus_cnt = t_convocation_newbie_exodus[['first_convocation_no','lower_name']].groupby(['first_convocation_no'],as_index = False).agg({'lower_name':'count'}).rename(columns={'lower_name':'lower_name_cnt'}).sort_values(by=['first_convocation_no'], ascending = True)
t_convocation_newbie_exodus_cnt
```

```python
# то же - в долях от всех новичков данного созыва

t_convocation_newbie_exodus_ratio = pd.merge(t_convocation_newbie_exodus_cnt, t_convocation_newbie, how='inner', left_on=['first_convocation_no'], right_on = ['first_convocation_no']).rename(columns={'lower_name_cnt_x': 'newbie_exodus_cnt', 'lower_name_cnt_y': 'newbie_cnt', 'first_convocation_no':'convocation_no'})
t_convocation_newbie_exodus_ratio['newbie_exodus_ratio'] = t_convocation_newbie_exodus_ratio.newbie_exodus_cnt / t_convocation_newbie_exodus_ratio.newbie_cnt
t_convocation_newbie_exodus_ratio
```

```python
# СВОДНАЯ ВСЕХ МЕТРИК СОЗЫВОВ

t_convocation_metrics = pd.merge(t_popular_convocations, t_nomination_newbie_ratio[['convocation_no','newbie_nomination_cnt','newbie_nomination_ratio']], how='left', left_on=['convocation_no'], right_on = ['convocation_no']) 
t_convocation_metrics = pd.merge(t_convocation_metrics, t_convocation_size, how='left', left_on=['convocation_no'], right_on = ['convocation_no']) 
t_convocation_metrics = pd.merge(t_convocation_metrics, t_convocation_newbie_ratio1[['first_convocation_no','newbie_cnt','newbie_ratio']], how='left', left_on=['convocation_no'], right_on = ['first_convocation_no']) 
t_convocation_metrics = pd.merge(t_convocation_metrics, t_convocation_exodus_ratio1[['last_convocation_no','exodus_cnt','exodus_ratio']], how='left', left_on=['convocation_no'], right_on = ['last_convocation_no']) 
t_convocation_metrics = t_convocation_metrics.drop(['first_convocation_no','last_convocation_no'], 1)
t_convocation_metrics = pd.merge(t_convocation_metrics, t_avg_year_no[['convocation_no','avg_year_no','avg_year_no_faculty','avg_year_no_diff']], how='left', left_on=['convocation_no'], right_on = ['convocation_no']) 
t_convocation_metrics = pd.merge(t_convocation_metrics, t_convocation_newbie_exodus_ratio[['convocation_no','newbie_exodus_cnt','newbie_exodus_ratio']], how='left', left_on=['convocation_no'], right_on = ['convocation_no']) 

t_convocation_metrics
```
