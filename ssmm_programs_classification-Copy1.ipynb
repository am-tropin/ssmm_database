```python
convocation_token_list = []
tokenize_regex = re.compile(r'([\w\-]+)', re.I)

for i in range(len(programs_df)):
    test_txt = programs_df['program_txt'][i]
    tokens = tokenize_regex.findall(test_txt)
    convocation_token_list += [str.lower(t) for t in tokens]
    
print(convocation_token_list)
```

```python
# немного отредактировать из-за -

# for w in convocation_token_list:
#     if w[0] == '-':
#         print(w, w[1:])
#         w = w[1:]
        
convocation_token_list = [w for w in convocation_token_list if w[0] != '-'] + [w[1:] for w in convocation_token_list if w[0] == '-']

print(convocation_token_list)
```

```python
convocation_token_dict = {}
for w in convocation_token_list:
    convocation_token_dict[w] = {'type': '', 'parent': '', 'cnt': 0}
    
for w in convocation_token_list:
    convocation_token_dict[w]['cnt'] += 1

convocation_token_dict = dict(sorted(convocation_token_dict.items(), key=lambda item: item[1]['cnt'], reverse=True))
convocation_token_dict
```

```python
def part_of_speech(convocation_token_list):
    for w in convocation_token_list:
        if w in prepositions_list:
            convocation_token_dict[w]['type'] = 'preposition'
        elif w in conjunctions_list:
            convocation_token_dict[w]['type'] = 'conjunction'
        elif w in pronouns_list:
            convocation_token_dict[w]['type'] = 'pronoun'
        elif w in adverbs_list:
            convocation_token_dict[w]['type'] = 'adverb'
        elif w in particles_list:
            convocation_token_dict[w]['type'] = 'particle'
        elif w in garbage_list:
            convocation_token_dict[w]['type'] = 'garbage'
        else:
            convocation_token_dict[w]['type'] = 'meaningful'
    return convocation_token_dict

part_of_speech(convocation_token_dict)
```

```python

def full_list_of_tokens_for_branch(convocation_token_dict, branch_parent_tokens):
    full_list_of_tokens = []
    for w in convocation_token_dict:
        if convocation_token_dict[w]['parent'] in branch_parent_tokens:
            full_list_of_tokens.append(w)
    return full_list_of_tokens


cultmass_tokens = ['мероприятие', 'праздник', 'пифагор', 'конкурс', 'посвящение', 'культурно-массовый', 'посвятой']
sport_tokens = ['спорт']

rights_fight_tokens = ['решение', 'решить', 'проблема', 'самоуправление']

education_tokens = ['образование']
infrastructure_tokens = ['бытовой', 'проживание', 'макулатура']
inform_tokens = ['оповещение', 'оповещать', 'информирование']



full_cultmass_tokens = full_list_of_tokens_for_branch(meaningful_convocation_token_dict, cultmass_tokens)
full_sport_tokens = full_list_of_tokens_for_branch(meaningful_convocation_token_dict, sport_tokens)
full_rights_fight_tokens = full_list_of_tokens_for_branch(meaningful_convocation_token_dict, rights_fight_tokens)
full_education_tokens = full_list_of_tokens_for_branch(meaningful_convocation_token_dict, education_tokens)
full_infrastructure_tokens = full_list_of_tokens_for_branch(meaningful_convocation_token_dict, infrastructure_tokens)
full_inform_tokens = full_list_of_tokens_for_branch(meaningful_convocation_token_dict, inform_tokens)

# удалить неоднозначные варианты
full_rights_fight_tokens.remove('решил',)
full_rights_fight_tokens.remove('решила')


print('full_cultmass_tokens:', full_cultmass_tokens)
print('full_sport_tokens:', full_sport_tokens)
print('full_rights_fight_tokens:', full_rights_fight_tokens)
print('full_education_tokens:', full_education_tokens)
print('full_infrastructure_tokens:', full_infrastructure_tokens)
print('full_inform_tokens:', full_inform_tokens)
```
