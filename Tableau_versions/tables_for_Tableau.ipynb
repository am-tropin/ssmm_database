# 2. Applications table

```python
ssmm_elections_candidates_1_11 = pd.read_csv('../library: main/ssmm_elections_candidates_1_11.csv', sep='\t', encoding='utf-8')
ssmm_elections_candidates_1_11.head()
```

```python
ssmm_dict_year_stream = pd.read_csv('../dictionaries/ssmm_dict_year_stream.csv', sep='\t', encoding='utf-8')
# ssmm_dict_year_stream.head()
ssmm_dict_graduated_years = pd.read_csv('../dictionaries/ssmm_dict_graduated_years.csv', sep='\t', encoding='utf-8')
# ssmm_dict_year_stream.head()
ssmm_dict_convocations = pd.read_csv('../dictionaries/ssmm_dict_convocations.csv', sep='\t', encoding='utf-8')
# ssmm_dict_convocations.head()
```

```python
ys_df = ssmm_elections_candidates_1_11.merge(ssmm_dict_year_stream)
ys_df = ys_df.merge(ssmm_dict_graduated_years)
ys_df['program_txt_2'] = ys_df['program_txt']
ys_df = ys_df.drop(['program_txt'], axis=1).rename(columns={'program_txt_2':'program_txt'})
# ys_df.head()

ys_df = ys_df.replace(r'\n', ' / ', regex=True) 
ys_df.head()
```

```python
ssmm_programs_info_conv_1_11 = pd.read_csv('../library: main/ssmm_programs_info_conv_1_11.csv', sep='\t', encoding='utf-8')
ssmm_programs_info_conv_1_11.head()
```

```python
ssmm_elections_streams_1_11 = pd.read_csv('../library: main/ssmm_elections_streams_1_11.csv', sep='\t', encoding='utf-8')
ssmm_elections_streams_1_11 = ssmm_elections_streams_1_11[['convocation_no','year_stream_no','stream_size','voted_cnt','NOTA_cnt','invalid_cnt','quota']].drop_duplicates()
ssmm_elections_streams_1_11 #.head(60)
```

```python
pr_df = ssmm_programs_info_conv_1_11.drop(['program_txt'], axis=1)

ys_df = ys_df.merge(pr_df, left_on='conv_x_name', right_on='conv_x_name', how='outer')
# ys_df.head()
ys_df = ys_df.merge(ssmm_dict_convocations[['convocation_no','academic_year']], left_on='convocation_no', right_on='convocation_no', how='outer')
ys_df = ys_df.merge(ssmm_candidates_core[['name','gender']], left_on='name', right_on='name', how='outer')
# ys_df.head()
ys_df = ys_df.merge(ssmm_elections_streams_1_11[['convocation_no','year_stream_no','stream_size','voted_cnt','NOTA_cnt','invalid_cnt','quota']], left_on=['convocation_no','year_stream_no'], right_on=['convocation_no','year_stream_no'], how='left')
ys_df.head()
```

```python
# add number of becoming a member (by person)

test_df = ys_df[['name','convocation_no','elected_flg']].sort_values(['name','convocation_no'],
              ascending = [True,True]).reset_index().drop(['index'], axis=1)
test_df['elected_no'] = test_df[test_df['elected_flg'] == 1].groupby(['name']).cumcount()+1
test_df['elected_no'] = test_df['elected_no'].fillna(0).astype(int)
test_df['nominated_no'] = test_df.groupby(['name']).cumcount()+1
test_df.head()
```

```python
ys_df = ys_df.merge(test_df[['convocation_no','name','elected_no','nominated_no']], left_on=['convocation_no','name'], right_on=['convocation_no','name'], how='left')
ys_df.head()
```

```python
test_df_2 = ys_df[['convocation_no','year_stream_no','votes_cnt','name']]
test_df_2 = test_df_2.drop_duplicates().sort_values(['convocation_no','year_stream_no','votes_cnt'],
              ascending = [True,True,False]).reset_index().drop(['index'], axis=1)
test_df_2['elected_rank'] = test_df_2.groupby(['convocation_no','year_stream_no'])["votes_cnt"].rank("dense", ascending=False)
test_df_2['elected_num'] = test_df_2[test_df_2['votes_cnt'].notna()].groupby(['convocation_no','year_stream_no']).cumcount()+1
# test_df_2['elected_no'] = test_df_2['elected_no'].fillna(0).astype(int)
test_df_2[130:140]
```

```python
ys_df = ys_df.merge(test_df_2, left_on=['convocation_no','year_stream_no','votes_cnt','name'], right_on=['convocation_no','year_stream_no','votes_cnt','name'], how='left')
ys_df.head()
```

```python
# добавить техническую строчку для задания типов полей

list_of_columns = list(ys_df)
techical_dict_applications = {}

for key in list_of_columns:
    techical_dict_applications[key] = techical_dict[key]

# pd.DataFrame(techical_dict.items())
techical_df = pd.DataFrame([techical_dict_applications])
techical_df
```

```python
ys_df_2 = pd.concat([techical_df,ys_df])
ys_df_2.head()
```
